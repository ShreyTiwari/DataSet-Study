```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ds = pd.read_csv("Population.csv")

plt.hist(ds.AHS, bins=8)
plt.xlabel('Average Household Size')
plt.ylabel('Frequency')
plt.title('Average Household Size Histogram:\n[Almost Symmetric]')
plt.show()

print(np.mean(ds.AHS))
print(np.median(ds.AHS))

print('\nThe graph is mostly normal. We observe that the average household size is 3.\
\n[Thus zipcodes do not correspond to any part of india ;)]')

print('\nNote : Average Household Size Zero has a certain frequency,\nthis pertains to the unhabited areas.')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ds = pd.read_csv('Population.csv')

bp = plt.boxplot(ds.TM)
print('Outliers : ', bp['fliers'][0].get_ydata())
print('Lower whiskers : ',bp['whiskers'][0].get_ydata())
print('Upper whiskers : ',bp['whiskers'][1].get_ydata())
plt.ylabel('Total Number of Males')
plt.title('Distribution')
plt.show()

bp1 = plt.boxplot(ds.TF)
print('Outliers :',bp1['fliers'][0].get_ydata())
print('Lower whiskers : ',bp1['whiskers'][0].get_ydata())
print('Upper whiskers : ',bp1['whiskers'][1].get_ydata())
plt.ylabel('Total Number of Females')
plt.title('Distribution')
plt.show()

#Extract densely populated regions
print('We see that the outliers represent relatively more populated areas.\nThese areas could possibly be major cities or townships.')
print('\nThe Zip Codes of such areas are listed below:\n')
for i,row in ds.TM.iteritems():
    if(row>bp['whiskers'][1].get_ydata()[1]):
        print(ds.ZC.loc[i])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

ds = pd.read_csv('Population1.csv')

#To find out ratio of men to women.
mtow = ds.TM/ds.TF

ds = ds.assign(MTW = mtow)
#print(ds.columns)
#print('\n')

ds.to_csv('Population_extras.csv')

plt.plot(ds.ZC, ds.MTW)
plt.xlabel('Zip Code')
plt.ylabel('Ratio of Men to Women')
plt.title('Line Plot')
plt.show()

print('From the above plot we observe two things :')
print('\t(I)The gaps in between represent division by zero error or in other terms unhabited/sparsly habited areas.')
print('\t(II)The abrupt peaks may indicate errors in the data collection process\n')

print('Mean Ratio : ',np.mean(mtow))
print('\nAs the mean ratio is close to one the scatter plot must exhibit strong relation\nwith average slope equal to the mean ratio.')

plt.scatter(ds.TM, ds.TF)
plt.xlabel('Total Males In An Area')
plt.ylabel('Total Females In An Area')
plt.title('Scatter Plot')
plt.show()

print('\nFrom the figure above it is clear that the plot is in consistance with the predictions')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

ds = pd.read_csv('Population.csv')

plt.hist(ds.MA, bins=30)
plt.xlabel('Median Age')
plt.ylabel('Frequency')
plt.title('Median Age Histogram\n[Mostly Symmetric]')
plt.show()

print('\nThe plot represents the distribution of median ages across various areas under the census.')
print('The plot is a normal curve with a few outliers that are accounted for considering the unhabited/sparsly populated areas.')
print('The plot below helps in better visualisation')

grid = sns.FacetGrid(ds)
grid.map(sns.distplot, 'MA')
plt.show()

sns.boxplot(ds.MA, orient='v')
plt.show()
print('The mosty symmetric nature of the boxplot points towards the fact that the median ages \
across the zipcodes is evenly\ndistributed.')
print('We can also conclude from the histogram as well as the boxplot that the median and the mean of the median ages must lie\
\nvery close to each other (provided the outliers are taken care of).')

print('\n')

print('The mean of the median ages is : ',np.mean(ds.MA))
print('The median of median ages is : ',np.median(ds.MA))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

ds = pd.read_csv('Population.csv')

print('Lets Have A Look At A Few Graphs :\n')

plt.scatter(ds.TP, ds.TH)
plt.xlabel('Total Population')
plt.ylabel('Total Number of Households')
plt.title('Linear Relationship\n[Expected]')
plt.show()
print('The above graph doesnt surprise us as by logic we can tell that there be and co-relation between\
\nthe total pouplation and the total number of houses.')

plt.scatter(ds.TH, ds.AHS)
plt.xlabel('Total Households')
plt.ylabel('Average Household Size')
plt.title('Unrelated\n[Expected]')
plt.show()
print('This graph is expected as we wouldnt expect any relationship between total number of houses and\
\naverage household size.')

plt.scatter(ds.TP, ds.MA)
plt.xlabel('Total Population')
plt.ylabel('Median Age')
plt.title('Interesting !!!')
plt.show()
print('However, The above graph is interesting as it goes against the typical logic that the median age musnt\
\nshow any noticeable corelation with the total poulation of the area.')
print('The most plausible reason being the youths attraction towards major/famous cities, presence of certain\
\nuniversities or both. Either way the change is slight in nature and doesnt pose an consequence any which way.')

print('\nNote : Large variations in median age are noticed in regions with lower population.\
\nThis is because of the presence of fewer entities in the calculation of the median with each entity having a greater\
\neffect on the median value than usual.')
```
